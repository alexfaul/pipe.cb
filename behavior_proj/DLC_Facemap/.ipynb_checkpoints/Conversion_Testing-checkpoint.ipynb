```python
def areaCalc(row):
    # Vertical Distance
    vert1x = (float(row["5"]) - float(row["2"]))**2
    vert1y = (float(row["6"]) - float(row["3"]))**2
    
    d1 = math.sqrt(vert1x + vert1y)
    
    # Horizontal Distance
    vert2x = (float(row["11"]) - float(row["8"]))**2
    vert2y = (float(row["12"]) - float(row["9"]))**2
    
    d2 = math.sqrt(vert2x + vert2y)
    
    # Finisher
    
    area = math.pi * (d1 / 2) * (d2 / 2)
    
    return area
    
```

```python
dlc_path = "/Users/drewmichelini/Fall2020/DLC_Facemap/Sut3_191120_001_eyeDLC_resnet50_drew_Sut3_191117_redoApr9shuffle1_300000.csv"
dlc_data = pd.read_csv(dlc_path)

test1 = dlc_data.drop(dlc_data.index[0])
test2 = test1.drop(test1.index[0])
test2.columns = [str(x) for x in range(1,17)]
test2["area"] = test2.apply(areaCalc, axis=1)
test2





```
